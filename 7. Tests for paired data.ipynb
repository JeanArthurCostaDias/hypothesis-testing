# Paired Data Analysis

Paired data analysis involves comparing observations between two related groups, such as measurements taken on the same subjects under different conditions or before and after a treatment.

## Assumptions:
1. **Dependent Samples:** The pairs of observations are related or matched in some way.
2. **Normality:** The differences between paired observations should be approximately normally distributed.

## Paired t-Test:
The paired t-test is commonly used to compare the means of paired data. The test statistic is calculated as:

$$ t = \frac{\bar{d}}{\frac{s_d}{\sqrt{n}}} $$

where:
- $\bar{d}$ is the mean of the differences.
- $s_d$ is the standard deviation of the differences.
- $n$ is the number of paired observations.

The mean of the differences ($ \bar{d} $) is calculated as:

$$ \bar{d} = \frac{\sum_{i=1}^{n} (B_i - A_i)}{n} $$

The standard deviation of the differences ($ s_d $) is calculated as:

$$ s_d = \sqrt{\frac{\sum_{i=1}^{n} (B_i - A_i - \bar{d})^2}{n-1}} $$

The t-statistic follows a t-distribution with $n-1$ degrees of freedom.

## Example Data:

| Subject | Condition A | Condition B | Difference (B - A) |
|---------|-------------|-------------|---------------------|
| 1       | 10          | 15          | 5                   |
| 2       | 12          | 14          | 2                   |
| 3       | 8           | 10          | 2                   |
| 4       | 14          | 16          | 2                   |
| 5       | 9           | 11          | 2                   |

```python
import numpy as np
from scipy import stats
from mystats.plots.qqplot import qq_plot

d = np.array([9, -4, 2, 7, 9, -2])
n = len(d)

qq_plot(d)
```

State the following Hypothesis:  
$$H_0:  \bar{d} = 0$$
$$H_1: \bar{d} \geq 0$$  

And set significance level $\alpha$ = 0.05

```python
# Calculate mean and variance using the formula
mean_d = np.mean(d)
n = len(d)
df = n - 1

mu_0 = 0
alpha = 0.05

# Calculate variance using np.var with ddof=1
s_d = np.std(d, ddof=1)

t_statistic = (mean_d - mu_0)/(s_d/np.sqrt(n))
p_value = 1 - stats.t.cdf(t_statistic, df=df)
t_critical = stats.t.ppf(1-alpha, df=df)
print(f't-statistic: {t_statistic: 3.4f} \np-value: {p_value: 3.4f} \nt-critical: {t_critical: 3.4f}')
print('Reject H0:', p_value < alpha)
```
